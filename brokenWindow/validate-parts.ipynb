```python
import pandas
data_frame = pandas.read_csv("travistorrent_11_1_2017.csv", usecols=['tr_build_id', 'tr_job_id', 'tr_status', 'tr_log_status', 'tr_prev_build', 'gh_project_name'])
```

```python
passed = data_frame[data_frame.tr_status == "passed"]
failed = data_frame[data_frame.tr_status != "passed"]
print(passed.count(), failed.count())
print()

failed_slice = failed.tr_build_id
res = passed.where(lambda x: x['tr_prev_build'].isin(failed_slice))
res = res.dropna()
print(res.count())
res.head(500)
len(res.tr_build_id.unique())
```
